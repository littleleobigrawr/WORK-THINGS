## Example of showing tickets complete per month by product

- Can help spot where knowledge silos lie within the team with high product number.
- Help to see who needs support for which product with low product numbers. 
- Identify any products that were impacted at certain dates, corrolations can easily be drawn from incidents, updates and other events. 

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("product-dummy-data - Sheet1 (1).csv")

x = data["DAY"]
y1 = data["TIQ"]
y2 = data["IDU"]
y3 = data["RMS"]
y4 = data["SC"]
y5 = data["SendIt"]
y6 = data["Other"]


plt.plot(x, y1, color='red', label='Trace IQ', marker='x', linewidth=1)
# plt.plot(x, y2, color='pink', label='IDU', marker='x', linewidth=1)
# plt.plot(x, y3, color='green', label='RMS', marker='x', linewidth=1)
# plt.plot(x, y4, color='blue', label='Smartcleanse', marker='x', linewidth=1)
plt.plot(x, y5, color='black', label='Sendit', marker='x', linewidth=1)
plt.plot(x, y6, color='grey', label='Other', marker='x', linewidth=1)

plt.title("Total tickets closed by product through July 2024")

plt.xlabel("Day of the month")
plt.ylabel("Ticket count")
plt.grid()

plt.legend()
plt.show()
```

## Example of showing tickets complete per month by time

- Can see if high ticket number per month are due to simple tickets. Identifying this can help support with more complex tickets. 
- Understand certain trends if expected ticket sub types are taking longer than expected or shorter than expected. 
- Gives an insight to who is good at handling more complex tickets and who is proficient at simpler tickets. 

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("product-dummy-data - Sheet1 (1).csv")

categories = [
    'tickets completed within 2 hours',
    'tickets completed within 10 hours',
    'tickets completed within 24 hours',
    'tickets completed within 48 hours',
    'tickets completed over 48 hours'
]

total_counts = [
    data[cat].sum() for cat in categories
]

plt.figure(figsize=(10, 6))

bars = plt.bar(categories, total_counts, color=['red', 'pink', 'green', 'blue', 'black'])

plt.title("Total Tickets Completed by Time Frame through July 2024")
plt.xlabel("Time Frame")
plt.ylabel("Total Ticket Count")
plt.xticks(rotation=45, ha='right') 

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')

plt.tight_layout()
plt.show()
```

## Example showing ticket volume by Type and Product

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv("product-dummy-data - Sheet1 (1).csv")

# Define the categories
categories = ["Password Reset", "Error", "Incident", "Update Account", "IP Enable / Disable"]

# Handle NaN values: Remove or fill NaNs in the relevant columns
for category in categories:
    data[category].fillna(0, inplace=True)  # Option 1: Replace NaNs with 0
    # Alternatively, you could use: data[category].dropna(inplace=True) # Option 2: Drop NaN values

# Create histograms for each category
plt.figure(figsize=(14, 8))

# Plot histogram for each category
for category in categories:
    plt.hist(data[category], bins=5, alpha=0.5, label=category)

plt.title("Distribution of Ticket Counts by Category")
plt.xlabel("Ticket Count")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
